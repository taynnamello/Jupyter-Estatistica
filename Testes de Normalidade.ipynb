Testes de Normalidade

```python
from scipy import stats 
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Criação de uma váriavel com dados em uma distribuição normal com a função rvs (100 elementos)
dados = norm.rvs(size = 1000)
dados
```

```python
# histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
# Geração de gráficos para verificar se a distribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
# Execução de testes de Shapiro Wilk
# Segundo argumento é o valor de p, não há como rejeitar a hipótese nula
stats.shapiro(dados)
```

Dados não normais

```python
dados2 = skewnorm.rvs(4, size=1000)
```

```python
# histograma
plt.hist(dados2, bins =20)
plt.title('Dados')
```

```python
stats.shapiro(dados2)
```
